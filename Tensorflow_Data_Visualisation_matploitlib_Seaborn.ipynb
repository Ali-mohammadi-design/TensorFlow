**Seaborn**

Note: Seaborn is based on the matplotlib.

```python
df=pd.read_csv('/content/heart.csv')
```

```python
import seaborn as sns
```

Note: with distplot you can get the distribution of data.

```python
sns.distplot(df['age'])
```

Note: in this image you have the age data based on the counts for each age.

If we put kde on false we can have the better figure.

```python
sns.distplot(df['age'], kde=False)
```

Note: As is clear in the previous picture we have the age range from before and after that specific age. We can increase the number of range ages with bins.Look at the following examples.

```python
sns.distplot(df['age'], kde=False, bins=2)
```

```python
sns.distplot(df['age'], kde=False, bins=20)
```

```python
sns.distplot(df['age'], kde=False, bins=50)
```

Note: As is clear in the previous image on some ranges we do not have any specifc data.

Note: we can change the size of the figures also.

Note: you can also change the clors.

```python
plt.figure(figsize=(12,8))
sns.distplot(df['age'], kde=False, bins=50, color='red')
```

Note: you can also you countplot to count the number of the data.

```python
sns.countplot(x='sex', data=df)
```

Note: in this function you can also add another catogorical features on one specific data.

```python
sns.countplot(x='sex', data=df, hue='target')
```

Note: you can change the colors with palette. You can use various range of colors on it automatically.

```python
sns.countplot(x='sex', data=df, hue='target', palette='terrain')
```

**Box Plot**


With box plot we would show the distribution of the categories. in every box plot we would have four parts. First 25%, second 25%, third 25%, and last 25%.

Thus the 50% of the data is on the half and in the box.

In the middle of the box we have the midian which is the 50%.

Note: in boxplot the x should be categorial and y should be continuous feature.

```python
sns.boxplot(x='sex',y='age', data=df, palette='terrain' )
```

Note: as is clear on the previous picture. Sex 1 on average is younger than sex0

```python
sns.boxplot(x='sex',y='age', data=df, palette='terrain', hue='target' )
```

**scatterplot**

```python
sns.scatterplot(x='chol', y='trestbps', data=df)
```

Note: you always can add a hue to compare different categories on specifc category.

```python
sns.scatterplot(x='chol', y='trestbps', data=df, hue='sex', palette='Dark2')
```
