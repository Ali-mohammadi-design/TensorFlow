Note: Panads is a package that can organize data

```python
import numpy as np
import pandas as pd
```

```python
labels=['a','b','c']

mylist=[10,2,34]

arr=np.array([10,2,34])

d={'a':10, 'b':2, 'c':34}
```

Note: we can make organize this data as some series.

Every series include some data and index

```python
pd.Series(data=mylist, index=labels)
```

Note: we can also use arrays (in numpy) or dictionaries to make series.

```python
pd.Series(arr)
```

```python
pd.Series(d)
```

Note: We can grab information from the series

```python
saleq1=pd.Series(data=[23,45,7], index=['usa','japan','Germany'])
saleq2=pd.Series(data=[24,556,73], index=['usa','japan','Iraq'])
```

```python
saleq2['Iraq']
```

We can also call the number of the index and get the information

```python
saleq2[0]
```

Note: we can integrate panda's series

```python
saleq1+saleq2
```

Note: since germany and iraq were not in the both of the series we got NAN.

**DataFrame**

Note: dataframs are multiple series with the same index.

DataFrames has data, index, and also columns.

In a simple words, this is like an excell sheet.

```python
from numpy.random import randint
```

```python
data=randint(-100,100,(4,5))
data
```

```python
df=pd.DataFrame(data=data, index=['a','b','c','d'], columns=['usa','Canada','China','Iran','Turkey'])
df
```

Note: We can grab columns

```python
df['usa']
```

Note: We can grab multiple columns by mentioning a list

```python
df[['usa','Iran']]
```

Note: We can also add a new column

```python
df['new']=df['usa']+df['Iran']
df
```

Note: We can also remove a column.

```python
df.drop('new', axis=1)
```

Note: in drop function by default the axis is equal to 0 that means search on the rows. Thus it can not find the 'new' on the row and you get an error. Thus you should set the axis on 1 to be able to remove it.

```python
df.drop('a')
```

Note: in panda you can also grab the rows. However you have to use .loc[] to show that you are looking for the rows

```python
df.loc['b']
```

```python
df.loc[['a','b']]
```

Note: with iloc you can get the row with its number.

```python
df.iloc[0]
```

```python
df.iloc[0:3]
```

Note: you can also get some specific rows and columns

```python
df.loc['a','usa']
```

```python
df.loc[['a','b'],['usa','China']]
```
