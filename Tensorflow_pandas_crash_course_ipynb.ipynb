**Missing Data**

We have three ways dealing with missing data:

1- leave it as missing.


2-remove it.


3- fill it.

```python
df=pd.DataFrame({'A':[1,2,np.nan,4],
                 'B':[5,np.nan,np.nan,8],
                 'c':[10,20,30,40]
                 })
```

```python
df
```

Note: with drop.na() YOU CAN DROP ALL NAs

Note: by default it is on axis=0 (rows)

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

Note: By threshold you can specify the minimum number of non-na data.

```python
df.dropna(axis=1, thresh=1)
```

It says that if you have at least 1 non-na value you should keep it.

```python
df.dropna(axis=1, thresh=3)
```

with fillna you can fill the missing data. you can just specify the value that it should be filled with.

```python
df.fillna(value='NO DATA')
```

```python
df.fillna(value=0)
```

Note you can also use it for specific row or column

```python
df['A'].fillna(value=0)
```

```python
df['A']=df['A'].fillna(value=0)
```

```python
df
```

```python
df['B'].mean()
```

```python
df['B']=df['B'].fillna(value=df['B'].mean())
df
```

Note: you can also fill all the NAs with just a specific order like mean of every column

```python
df=pd.DataFrame({'A':[1,2,np.nan,4],
                 'B':[5,np.nan,np.nan,8],
                 'c':[10,20,30,40]
                 })
```

```python
df
```

```python
df.fillna(df.mean())
```

**Group by**

With groupby we can group similar data together and do some functions on them.

```python
df=pd.read_csv('/content/Universities.csv')
```

```python
df.head()
```

```python
df.groupby('Year')
```

Note: when you use groupby you have to do some aggregation operation on it. like mean, sum, std

```python
df.groupby('Year').sum()
```

```python
df.groupby('Year').mean()
```

Note: you can sort the information separated by group

```python
df.groupby('Year').sum().sort_index(ascending=True)
```

Note: We can group with two or more columns.

```python
df.groupby(['Year','Sector']).sum()
```

Note: You can use describe() to get the information of the data in the groupby().

```python
df.groupby('Year').describe()
```

Note: we can also transpose the matrix and change the position of the year and completion.

```python
df.groupby('Year').describe().transpose()
```

**Pandas Operation**

```python
df_one = pd.DataFrame({'k1':['A','A','B','B','C','C'],
                      'col1':[100,200,300,300,400,500],
                      'col2':['NY','CA','WA','WA','AK','NV']})
```

```python
df_one
```

```python
df_one['col2'].unique()
```

Note: nunique() would return the number of unique elements.

```python
df_one['col2'].nunique()
```

Note: you can also get the count of each element with value_counts()

```python
df_one['col2'].value_counts()
```

Note: you can also drop the duplicate with the following codes.

```python
df_one['col2'].drop_duplicates()
```

```python
df_one.drop_duplicates()
```

```python
def grab_first_letter(text):
  return text[0]
grab_first_letter("ali")
```

```python
df_one['col2'].apply(grab_first_letter)
```

```python
df_one['first_letter_Col2']=df_one['col2'].apply(grab_first_letter)
df_one
```

```python
my_map={'A':1, "B":2, "C":3}
```

```python
df_one['numbers']=df_one['k1'].map(my_map)
df_one
```
